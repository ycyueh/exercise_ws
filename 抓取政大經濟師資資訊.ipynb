```python
# 匯入套件
from bs4 import BeautifulSoup as bs
import requests as req
from pprint import pprint
```

```python
###
##抓取政大經濟師資研究領域與email
url = 'https://econo.nccu.edu.tw/PageStaffing?fid=3274'

res = req.get(url)
soup = bs(res.text,'lxml')
teacher_info = []
```

```python
teacher_info.clear()

for teacher in soup.select('section.page-staff-info'):
    name = teacher.select_one('strong[class="name"]').get_text()
    web = teacher.select_one('a')
    email = teacher.select_one('ul > li.info-mail >p >a')
    teacher_info.append({
        'name':name[0:3],
        'web':web["href"],
        'email':email["href"]
    })
   # print(strong.get_text())
pprint(teacher_info)
```

```python
### 走訪網頁，抓下研究專長
for index,web in enumerate(teacher_info):
    res = req.get(web['web'])
    soup = bs(res.text,'lxml')
    table = soup.select('div.staff-tb>table>tbody>tr')
    table = table[len(table)-1]
    t = table.find('td').text
    teacher_info[index]['研究專長'] = t

pprint(teacher_info)
```
