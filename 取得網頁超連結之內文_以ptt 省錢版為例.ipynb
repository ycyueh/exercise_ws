```python
# 匯入套件
from bs4 import BeautifulSoup as bs
import requests as req
from pprint import pprint
```

```python
##ptt省錢版
url = 'https://www.ptt.cc/bbs/Lifeismoney/index.html'

res = req.get(url)
##剖析 用lxml
soup = bs(res.text,'lxml')

###建立 list 儲存資訊
lists_post = []
```

#### css selector
接在 剖析過的soup後
語法：soup.select()

```python
lists_post.clear()
### a[herf]：選擇有herf屬性的a
for a in soup.select('div.r-ent > div.title > a[href]'):
    lists_post.append({
        'title':a.get_text(),
        'link':'https://www.ptt.cc'+a['href']
    })
    pprint(lists_post)
```

#### 用request再次走訪網頁，取得內文

```python
for index,cont in enumerate(lists_post):
    res2 = req.get(cont['link'])
    soup2 = bs(res2.text,'lxml')
    ##去掉標題作者時間
    #['屬性^="" '] whose 屬性 attribute value begins with "article-metaline"
    for div in soup2.select('div[class^="article-metaline"]'):
        div.decompose() ## 去除
    ##去除本文超連結(不一定有)
    if len(soup2.select('span.f2 > a[href][rel = "noreferrer.noopener.nofollow"]')):
        soup2.select_one('span.f2 > a[href][rel = "noreferrer.noopener.nofollow"]').decompose()
    ##去除span.f2
    for span in soup2.select('span.f2'):
        span.decompose()
    ##去除推文
    for div in soup2.select('div.push'):
        div.decompose()
    ## 取得實際需要的內文
    ## 此處的 soup2 已經是去除以上內容的內文
    html = soup2.select_one('div#main-content').decode_contents()

    ##把內文存乳list中
    ## 新增一欄為 content
    lists_post[index]['content']=html

pprint(lists_post)
```
