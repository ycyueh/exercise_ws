```python
# 使用 requests 工具
import requests

# 使用 json 工具
import json
```

```python
url = 'https://data.taipei/api/v1/dataset/8f690548-61bc-4bff-8baa-01d465eb672c?scope=resourceAquire'
# 使用 GET 方式下載普通網頁
res = requests.get(url)
print(res.status_code) # 伺服器回應的狀態碼
```

```python
print(res.encoding)
```

```python
# 輸出網頁 HTML 原始碼
print(res.text)
```

```python
data = json.loads(res.text)
```

```python
data
```

```python
print(data['result']['limit'])
print(data['result']['offset'])
```

```python
##取得部分節點文字
for obj in data['result']['results']:
    print(obj['_id'])
    print(obj['臺北市youbike每月使用量'])
    print(obj['數量（次數）'])
```
