```python
import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint
```

```python
url = 'http://www.atmovies.com.tw/movie/next/'
```

```python
res = req.get(url)
res.encoding = 'utf-8'
soup = bs(res.text,'lxml')
```

```python
soup.select('a')
movie_list=soup.select('ul')[5]
for name in movie_list.select('div.filmtitle'):
    print(name.get_text())
```

```python
for rundate in movie_list.select('div.runtime'):
    a_tag = rundate.find('a')
    if a_tag:
        print('上映日期：',a_tag.get_text(),sep='')
    else:
        print(rundate.get_text()[0:15])
    # if date.has_('a'):
    #     print(date.select_one('a').get_text())
   # print(date.get_text())
```

```python
##列表資訊
movie = []
```

```python
#先清空
movie.clear()
##取得電影名稱
movie_list=soup.select('ul')[5]
for name in movie_list.select('div.filmtitle'):
    print(name.get_text())
    #加入movie list
    movie.append({
        'title' :name.get_text()
    })
##上映日
for rundate in movie_list.select('div.runtime'):
    a_tag = rundate.find('a')
    if a_tag:
        rund = '上映日期：'+a_tag.get_text()
        print(rund)
    else:
        rund = rundate.get_text()[0:15]
        print(rund)
    movie.append({
        '上映日':rund
    })
```

```python
pprint(movie)
```

```python
import pandas as pd
pd.concat([pd.DataFrame.from_dict(movie[0:20]),
pd.DataFrame.from_dict(movie[20:40])],axis=1)
# movie
```
