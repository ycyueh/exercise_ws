```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.ptt.cc/bbs/Lifeismoney/index.html'
```

```python
res =req.get(url)
## lxml解析器
soup = bs(res.text,'lxml')
```

```python
soup.find('a')
```

```python

soup.find_all('div',class_='r-ent')
```

```python
##回傳所有列表標題和連結 
##title
print(soup.select_one('title'))
##只輸出text：.get_text()
print(soup.select_one('title').get_text())
```

```python
name = list()
```

```python

for i,a in enumerate(soup.select('a')):
    name[i] = a.get_text()
    #print(a.get_text())
```

```python
# 透過迭代取得所有 a 的屬性 href
for a in soup.select('a'):
    if a.has_attr('href'):
        print('https://www.ptt.cc/',a['href'],sep = '')
    else:
        print("=" * 50)
        print(f'連結[{a.get_text()}]沒有 href 屬性')
        print("=" * 50)
```
